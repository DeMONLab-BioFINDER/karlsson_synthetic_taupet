```python
#imports
import sys
from pathlib import Path
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

SRC_DIR = Path().resolve().parents[0] / "src"
sys.path.append(str(SRC_DIR))
import utils.image_helpers as ih
import utils.plot_helpers as ph
```

```python
## Config

# Save plots or not and specify save name
save_plots = True 
savename = 'unet_example_cases'

# Specify the file names of the 10 example cases to plot
names = ['']

# Specify the directory for input MRI and true tau-PET images
mri_root = ' '
taupet_root = ' '
```

```python
# load the MRI, synthetic PET and true PET images

mri_cases = []
pred_cases = []
true_cases = []

for name in names:
    mri_img = ih.load_and_preprocess_nifti(mri_root + name + '/jademarec.nii.gz')
    synt_img = nib.load('../outputs/synthetic_test_scans/' + name + '.nii.gz')
    synt_img_data = synt_img.get_fdata()

    true_img = nib.load(taupet_root + name + '/smoothed_image.nii.gz')
    true_img_data = true_img.get_fdata()

    mri_cases.append(mri_img[:,:,:,0])
    pred_cases.append(synt_img_data)
    true_cases.append(true_img_data)

mri_cases = np.array(mri_cases)
pred_cases = np.array(pred_cases)
true_cases = np.array(true_cases)
```

```python
# Plot the images as in Figure 5 in the article

plt.rc('font', size=10)
figure, axs = plt.subplots(3,10,figsize=(15,4.5),sharey=False,sharex=False)
plt.subplots_adjust(wspace = 0.1,hspace=0.1)

cmap = ph.get_cmap_pet()

loc = 45
rot = 90
for j in range(len(mri_cases)):
    axs[0,j].imshow(ndimage.rotate(mri_cases[j,:,loc,:],rot), vmin=-1.5,vmax=1.5,cmap='gray')
    axs[1,j].imshow(ndimage.rotate(pred_cases[j,:,loc,:],rot), vmin=0.5,vmax=3,cmap=cmap)
    axs[2,j].imshow(ndimage.rotate(true_cases[j,:,loc,:],rot), vmin=0.5,vmax=3,cmap=cmap)
    for i in range(3):
        axs[i,j].set_xticks([])
        axs[i,j].set_yticks([])
    
axs[0,0].set_ylabel('MRI')
axs[1,0].set_ylabel('Synthetic PET')
axs[2,0].set_ylabel('True PET')

if save_plots:
    plt.savefig('../outputs/figures/' + savename + '.pdf',bbox_inches='tight')
```
